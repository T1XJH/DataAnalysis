```python
def make_df(cols, inds):
    data = {c:[c+str(i) for i in inds]for c in cols}
    return DataFrame(data, index = inds, columns = cols)
```

```python
make_df(['A','B'],[1,2])
```

```python
df1 = make_df(list('AB'),[0,1])
df2 = make_df(list('AB'),[2,3])
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3 = make_df(list('AB'),[0,1,2])
df4 = make_df(list('AB'),[1,2,3])
```

```python
df5 = pd.concat((df3, df4))
df5
```

```python
df5.loc[[1,2]]
```

```python
df1 = make_df(list('AB'),[1,3])
df2 = make_df(list('BC'),[2,4])
```

```python
# join_axes以某一个DataFrame列索引为新的索引值
pd.concat([df3,df4],join_axes=[df3.columns])
```

```python
# merge合并
df1 = DataFrame({'employee': ['Po','Sara','Danis'],
                 'group': ['sail','counting','marketing']})
df2 = DataFrame({'employee': ['Po','Sara','Bush'],
                 'work_time': [2,3,1]})
display(df1, df2)
```

```python
#一对一合并
pd.merge(df1, df2)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1,df2], axis = 1)
```

```python
df3 = DataFrame({'employee': ['Po','Summer','Flower'],
                 'group': ['sail','marketing','serch'],
                 'salary': [12000, 10000, 8000]})
df4 = DataFrame({'employee': ['Po','Winter','Flower'],
                 'group': ['marketing','marketing','serch'],
                 'work_time': [2, 1, 5]})
display(df3, df4)
```

```python
pd.merge(df3, df4)
```

```python
pd.merge(df3,df4,on = 'employee')
```

```python
pd.merge(df3,df4,on = 'group', suffixes=['_A','_B'])
```

```python
pd.merge(df3, df4, left_on = 'employee', right_on='employee')
```

```python
df3 = DataFrame({'employer': ['Po','Summer','Flower'],
                 'Team': ['sail','marketing','serch'],
                 'salary': [12000, 10000, 8000]})
df4 = DataFrame({'employee': ['Po','Winter','Flower'],
                 'group': ['marketing','marketing','serch'],
                 'work_time': [2, 1, 5]})
display(df3, df4)
```

```python
pd.merge(df3, df4, left_on = 'employer', right_on='employee')
```

```python
pd.merge(df3, df4, left_on = 'Team', right_on='group')
```
