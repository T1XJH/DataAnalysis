```python
#排序
n1 = np.array([2,5,1,7,4])
```

```python
n1.size
```

```python
def sortn(nd):
    for i in range(nd.size):
        for j in range(i, nd.size):
            if nd[i] > nd[j]:
                nd[i],nd[j] = nd[j],nd[i]
    return nd
```

```python
sortn(n1)
```

```python
n1 = np.array([2,5,1,7,4])
```

```python
# 降低运算的空间复杂度和时间复杂度
def sortnd(nd):
    for i in range(nd.size):
        min_index = np.argmin(nd[i:])+i
#         print(min_index)
        nd[i],nd[min_index] = nd[min_index],nd[i]
    return nd
```

```python
sortnd(n1)
```
